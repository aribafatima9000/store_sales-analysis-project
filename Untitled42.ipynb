# Discount Effectiveness

```python
# Group by DiscountApplied and PreviousPurchases
grouped = df.groupby(['DiscountApplied(%)','PreviousPurchases']).agg({
    'ItemRating':'mean',   # Average rating
    'Amount':'mean'        # Average purchase amount
}).reset_index()

print(grouped)
```

```python
import matplotlib.pyplot as plt

# Pivot table for plotting
pivot_rating = grouped.pivot(index='PreviousPurchases', columns='DiscountApplied(%)', values='ItemRating')
pivot_amount = grouped.pivot(index='PreviousPurchases', columns='DiscountApplied(%)', values='Amount')

# Plot ItemRating
pivot_rating.plot(kind='bar', figsize=(10,5))
plt.title("Effect of Discount & Previous Purchases on Item Rating")
plt.xlabel("Previous Purchases")
plt.ylabel("Average Item Rating")
plt.legend(
    title='Discount (%)',       # Legend title
    title_fontsize=12,          # Legend title font size
    fontsize=10,                # Legend labels font size
    loc='right',           # Position: 'upper right', 'lower left', etc.
    frameon=True,               # Add border box
    shadow=True                 # Add shadow to legend
)
plt.show()

# Plot Amount
pivot_amount.plot(kind='bar', figsize=(10,5))
plt.title("Effect of Discount & Previous Purchases on Purchase Amount")
plt.xlabel("Previous Purchases")
plt.ylabel("Average Purchase Amount")
plt.legend(
    title='Discount (%)',       # Legend title
    title_fontsize=12,          # Legend title font size
    fontsize=10,                # Legend labels font size
    loc='right',           # Position: 'upper right', 'lower left', etc.
    frameon=True,               # Add border box
    shadow=True                 # Add shadow to legend
)
plt.show()
```

# High-Value Customer Identification

```python


customer_stats = df.groupby('CustomerID')['Amount'].agg(
    mean='mean',
    std='std',
    transaction_count='count'
).reset_index()

# Replace NaN std (single transaction customers)
customer_stats['std'] = customer_stats['std'].fillna(0)

# ==============================
# 3. Define High-Value Consistent Buyers
# ==============================

high_value_customers = customer_stats[
    (customer_stats['mean'] > 200) & 
    (customer_stats['std'] < 50) & 
    (customer_stats['transaction_count'] >= 3)
]

print("\n=== High Value Consistent Customers ===")
print(high_value_customers)

# ==============================
# 4. Merge with Original Data
# ==============================

patterns = pd.merge(
    high_value_customers[['CustomerID']], 
    df, 
    on='CustomerID'
)

print("\n=== Transaction Patterns ===")
print(patterns[['CustomerID','Age','Gender','Category','Amount']])

# ==============================
# 5. Pattern Analysis
# ==============================

print("\n=== Age Distribution ===")
print(patterns['Age'].describe())

print("\n=== Gender Distribution ===")
print(patterns['Gender'].value_counts())

print("\n=== Preferred Categories ===")
print(patterns['Category'].value_counts())

print("\nAnalysis Complete ")
```

# 6. Product Category Insights

```python


# Group by category and calculate average rating and amount
category_stats = df.groupby('Category').agg({
    'ItemRating':'mean',
    'Amount':'mean'
}).reset_index()

print(category_stats)

# Identify highest rating and purchase amount
highest_rating_category = category_stats.loc[category_stats['ItemRating'].idxmax()]
highest_amount_category = category_stats.loc[category_stats['Amount'].idxmax()]

print("\nHighest Average Rating Category:\n", highest_rating_category)
print("\nHighest Average Purchase Amount Category:\n", highest_amount_category)
```
